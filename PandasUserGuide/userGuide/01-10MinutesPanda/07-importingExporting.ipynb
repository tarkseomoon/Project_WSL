```python
idx = pd.date_range('01/01/2010', periods=1000)
csv_df = pd.DataFrame(
    np.random.randn(1000, 4),
    index = idx,
)

csv_df = csv_df.reset_index()
csv_df.columns=['date', 'A', 'B', 'C', 'D']
csv_df.to_csv('./foo.csv', index=False)
```
