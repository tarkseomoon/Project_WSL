```python
#  Plotting
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# plt.close(all)
```

```python
ts = pd.Series(np.random.randn(1000), index = pd.date_range('01/01/2020', periods=1000))
ts =ts.cumsum()
ts.plot();
```

```python
plt.show()
```

```python
df = pd.DataFrame(
    np. random.randn(1000, 4), index=ts.index, columns=['A', 'B', 'C', 'D']
    # 다른 DataFrame의 index를 빌려 올 수도 있다.
)

df = df.cumsum()
plt.figure();
df.plot();
plt.legend(loc='best');
```
