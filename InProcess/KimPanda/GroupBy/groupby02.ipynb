```python
pd.options.display.float_format = '{:.2f}'.format
a1 = ['송중기', '남', 1, 79, 93]
a2 = ['김나현', '여', 1, 85, 85]
a3 = ['권보아', '여', 1, 91, 78]
a4 = ['김범수', '남', 1, 97, 98]
a5 = ['이효리', '여', 2, 93, 79]
a6 = ['박효신', '남', 2, 78, 90]
a7 = ['한예슬', '여', 2, 87, 74]
a8 = ['민경훈', '남', 2, 99, 89]
col1 = ["이름", "성별", "반", "국어", "영어"]
df1 = pd.DataFrame([a1, a2, a3, a4, a5, a6, a7, a8], columns=col1)
df2 = df1.copy()
df2.loc["이효리", "반"] = 1
df2["인원"] = 1
s1 = df1["국어"].copy()
df1.drop(columns='이름', inplace=True)
df1
```

```python
upv_df = pd.melt(df1, id_vars=('성별', '반'),value_vars=('국어', '영어'))
upv_df.columns = ['성별', '반', '과목', '점수']
upv_df
```

```python
df1.groupby(by=['성별', '반']).agg(['mean'])[['국어', '영어']]
```

```python
pd.pivot_table(upv_df, index= ['성별', '반'], columns=['과목'], aggfunc='median')
```

```python
pvt_df1=df1.pivot_table(values=['국어','영어'], index='성별', columns='반', aggfunc='mean')
pvt_df1.stack()
```
