```python
idx1 = ["송중기", "김나현", "박효신"]
a1 = ['2021-01-01', '송중기', 10000]
a2 = ['2021-01-01', '김나현', 2000]
a3 = ['2021-01-02', '송중기', -5000]
a4 = ['2021-01-02', '김나현', 5000]
a5 = ['2021-01-03', '김나현', -3000]
a6 = ['2021-01-03', '송중기', 5000]
df1 = pd.DataFrame([[88,72], [78,93], [91,85]], index=idx1, columns=["국어","수학"])
df2 = df1.stack().reset_index()
df2.columns = ["이름", "과목", "점수"]
df3 = pd.DataFrame([2,3,4,5], columns = ["number"])
df4 = pd.DataFrame([a1, a2, a3, a4, a5, a6], columns=["날짜", "이름", "입출금"])
```

```python
display(df1, df1.rank(method = 'first', ascending=False), df1.rank(pct=True, method = 'first',  ascending=False))  # method는 group내 중복이 있을 때 처리한 방법을 정한다.
```

```python
df1 = df1.reset_index().rename(columns={'':'이름'})
df1
upv_df1 = pd.melt(df1, id_vars='index', value_vars = (['국어', '수학']), var_name='과목', value_name='점수')
upv_df1
```

```python
upv_df1['과목내 순위'] = upv_df1.groupby(by='과목')['점수'].rank(ascending=False).astype(int)
upv_df1
```

```python
df4['잔고'] = df4.groupby('이름')['입출금'].cumsum()
df4
```
