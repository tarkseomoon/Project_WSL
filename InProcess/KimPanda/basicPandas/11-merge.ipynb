```python
import numpy as np
import pandas as pd
```

```python
names1 = ['송중기', '권보아', '김나현', '박효신', '진달래']
a1, a2, a3, a4, a5 = '송중기', '권보아', '김나현', '박효신', '진달래'
b1, b2, b3 = '카페라테', '아이스티', '모카라테'
c1, c2 = '콩다방', '별다방'

df1 = pd.DataFrame([names1, [b1, b2, b3, b2, b3]], index=['이름', '제품']).T
df2 = pd.DataFrame([[b1, 4500],[b2, 4600], [b3, 4800]], columns=['제품', '가격'])
df3 = pd.DataFrame([names1, [c1, c1, c2, c1, c2], [b2, b1, b2, b2, b1]], index=['이름', '업체', '제품']).T
df4 = pd.DataFrame([[c1, b2, 4600], [c1, b1, 4700], [c2, b2, 4200], [c2, b1, 4300]], columns=['업체', '제품', '가격'])
df5 = pd.DataFrame([[b1, 4500], [b2, 4600], [b3, 4800]], columns=['상품', '가격'])
df6 = pd.DataFrame([[b1, 4500], [b2, 4600], [b3, 4800], [b3, 4800]], columns=['제품', '가격'])
df7 = pd.DataFrame([[a2, a3, a1, a5], [b2, b1, b1, b3]], index=['이름', '1일차']).T
df8 = pd.DataFrame([[a2, a3, a4, a1], [b1, b3, b2, b2]], index=['이름', '2일차']).T
df9 = pd.DataFrame([b1, b3, b2, b2], index=[a2, a3, a4, a1], columns=['2일차'])
# df9 = pd.Series([b1, b3, b2, b2], index=[a2, a3, a4, a1], name='2일차')
```
