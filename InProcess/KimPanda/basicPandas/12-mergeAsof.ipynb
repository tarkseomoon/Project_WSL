```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame([['송중기', 92], ['권보아', 89], ['김나현', 71], ['박효신', 90]], columns=['이름', '점수'])
df2 = pd.DataFrame([['A', 90], ['B', 80], ['C', 70]], columns=['학점', '점수'])
df3 = pd.DataFrame([['A', 100], ['B', 90], ['C', 80]],columns=['학점', '점수'])
df4 = pd.DataFrame([['A', 95], ['B', 85], ['C', 75]], columns=['학점', '점수'])
df5 = pd.DataFrame([[90, 'lucky'], [80, 'lucky']], columns=['점수', '구분'])

t1 = ['2021-02-11', '2021-02-12', '2021-09-21','2021-09-22', '2021-09-23']
# 그저 string의 list일뿐
df6 = pd.DataFrame([t1], index=['날짜']).T
df7 = pd.DataFrame([['2021-02-12', '설날'], ['2021-09-21', '추석']], columns=['날짜', '내용'])
df6['날짜'] = pd.to_datetime(df6['날짜'])
df7['날짜'] = pd.to_datetime(df7['날짜'])
# string을 TimestampSeries로 변환

df8 = pd.DataFrame([['송중기', '국어', 99], ['송중기', '영어', 72], ['권보아', '국어', 92], ['권보아', '영어', 92]],\
                   columns=['이름', '과목', '점수'])
df9 = pd.DataFrame([[90, 'A', '국어'], [0, 'B', '국어'], [95, 'A', '영어'], [0, 'B', '영어'], ],\
                   columns=['점수', '학점', '과목'])
```

```python
df1 = df1.sort_values('점수')
df2 = df2.sort_values('점수')
df3 = df3.sort_values('점수')
df4 = df4.sort_values('점수')
df5 = df5.sort_values('점수')
df6 = df6.sort_values('날짜')
df7 = df7.sort_values('날짜')
df8 = df8.sort_values('점수')
df9 = df9.sort_values('점수')
# left, right 공히 sort되어 있어야 한다. 
```

```python
pd.merge_asof(df1, df2)
# 공통으로 점수column을 갖고 있으므로 on=은 생략
```

```python
pd.merge_asof(df1, df2, allow_exact_matches=False)
# 자신보다 작은 수를 찾는다.
```

```python
pd.merge_asof(df1, df2, direction='forward')
```

```python
pd.merge_asof(df1, df3, direction='forward')
```

```python
pd.merge_asof(df1, df3, direction='forward', allow_exact_matches=False)
# 자신의 숫자보다 큰 수를 찾는다.
```

```python
df5
```

```python
pd.merge_asof(pd.merge_asof(df1, df2), df5, direction= 'backward', tolerance=2)
# direction이 backward이면 tolerance는 forward로 작용한다.
```

```python
pd.merge_asof(pd.merge_asof(df1, df2), df5, direction= 'forward', tolerance=2)
# direction이 forward이면 tolerance는 backward로 작용한다.
```

```python
pd.merge_asof(df6, df7, direction='forward', tolerance=pd.Timedelta('1d'))
```

```python
pd.merge_asof(df8, df9, on='점수', by='과목')
# by=과목는 과목별 group을 만들어 merge_asof처리 한다. groupby에서 왔다고 해서 by   🌷
```

```python
df = pd.DataFrame({
    'col1': ['A', 'A', 'B', np.nan, 'D', 'C'],
    'col2': [2, 1, 9, 8, 7, 4],
    'col3': [0, 1, 9, 4, 2, 3],
    'col4': ['a', 'B', 'c', 'D', 'e', 'F']
})

df.sort_values(by='col4', key=lambda col: col.str.lower())
# 본 작업에 들어가기 전에 Series에 대해 전처리(lambda)를 한다.
# neverthless, series의 original values는 보존된다.
```

```python
quotes = pd.DataFrame(
    {
        "time": [
            pd.Timestamp("2016-05-25 13:30:00.023"),
            pd.Timestamp("2016-05-25 13:30:00.023"),
            pd.Timestamp("2016-05-25 13:30:00.030"),
            pd.Timestamp("2016-05-25 13:30:00.041"),
            pd.Timestamp("2016-05-25 13:30:00.048"),
            pd.Timestamp("2016-05-25 13:30:00.049"),
            pd.Timestamp("2016-05-25 13:30:00.072"),
            pd.Timestamp("2016-05-25 13:30:00.075")
        ],
        "ticker": [
               "GOOG",
               "MSFT",
               "MSFT",
               "MSFT",
               "GOOG",
               "AAPL",
               "GOOG",
               "MSFT"
           ],
           "bid": [720.50, 51.95, 51.97, 51.99, 720.50, 97.99, 720.50, 52.01],
           "ask": [720.93, 51.96, 51.98, 52.00, 720.93, 98.01, 720.88, 52.03]
    }
)

trades = pd.DataFrame(
       {
           "time": [
               pd.Timestamp("2016-05-25 13:30:00.023"),
               pd.Timestamp("2016-05-25 13:30:00.038"),
               pd.Timestamp("2016-05-25 13:30:00.048"),
               pd.Timestamp("2016-05-25 13:30:00.048"),
               pd.Timestamp("2016-05-25 13:30:00.048")
           ],
           "ticker": ["MSFT", "MSFT", "GOOG", "GOOG", "AAPL"],
           "price": [51.95, 51.95, 720.77, 720.92, 98.0],
           "quantity": [75, 155, 100, 100, 100]
       }
   )
```

```python
quotes
```

```python
trades
```

```python
pd.merge_asof(trades, quotes, on="time", by="ticker")
# 기준이 되는 것은 left, left에 맞는 right를 찾아야 한다
```

```python
pd.merge_asof(
    trades, quotes, on="time", by="ticker", tolerance=pd.Timedelta("2ms")
)
# direction의 default가 backward이므로 지정 trade시간 이전에 해당 ticker가 있는지를 확인한다
# 048이전에 AAPL이 없으므로 bid, ask는 NaN이 된다.
```
