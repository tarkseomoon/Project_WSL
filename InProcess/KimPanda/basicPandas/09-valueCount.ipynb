```python
import pandas as pd
import numpy as np
```

```python
a1 = ['송중기', '김나연', '박보검', '권보아', '이효리', '김범수']
b1 = ['별다방', '콩다방', '별다방', '별다방', '별다방', '콩다방']
c1 = ['카페라떼', '아이스티', '카페라떼', '카페라떼', '아이스티', '카페라떼']

df1 = pd.DataFrame([a1, b1, c1], index =['이름', '업체', '제품']).T
df2 = pd.read_csv("https://raw.githubusercontent.com/"
                  "panda-kim/pandas/main/value_counts_01.csv")
```

```python
df2 = df2.to_csv('value_count.csv', index=False)
df2 = pd.read_csv('value_count.csv')
```

```python
df11 = df1.value_counts(['업체', '제품'], normalize=True).to_frame()
# normalize: proportion(비율) rather than frequencies(빈도), column명도 count에서 proposition으로 변경
df11
```

```python
df2.value_counts(subset=['반', '선택']).sort_index(level=[0, 1]).to_frame()
```

```python
index = df2.value_counts('선택').index[:3]
# value_counts([...])에서 [...]를 사용하게 되면, 하나의 인자만을 넣더라도 MultiIndex Series를 생성한게 된다.
# 따라서 index를 추출하면, tuple()형의 index를 얻게 되어 이를 단순 list-like로 만들어 isin을 쓰기 불편해진다.
```

```python
df2[df2['선택'].isin(index)]    # boolean indexing
```
