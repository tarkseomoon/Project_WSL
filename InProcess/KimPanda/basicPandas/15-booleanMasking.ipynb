```python
import numpy as np
import pandas as pd
```

```python
col1 = ['이름', '국어', '영어']
df1 = pd.DataFrame([['송중기', 77, 80], ['김나현', 92, 72], ['권보아', 62, 88], ['박효신', 91, 95]], columns=col1)
df2 = df1.copy()
df3 = df1.copy()
df4 = df1.copy()
df5 = df1.copy()
```

```python
df1.loc[df1['국어'] > 90, '국어'] = 'A'
df1
```

```python
df2['국어'] = df2['국어'].mask(df2['국어'] > 90, 'A')
df2
```

```python
df3['국어'].mask(df3['영어'] > 80, 'R/Out')
# df3['국어_Final'] = df3['국어'].mask(df3['영어'] > 80, 'R/Out')
```

```python
df3['Rule_Out'] =pd.Series() 
# df3['Rule_Out'] ="" 
df3['Rule_Out'] = df3['Rule_Out'].mask(df3['영어'] > 80, True)
df3
# mask() pd.Series에 붙어야 하므로, 먼저 Empty Series를 만들어 준다(두가지 방법)
```

```python
df4['Archivement']= ['A' if (i > 90) & (j >90) else 'B' if (i > 80) & (j >80) else 'C' for i, j in zip(df4.국어, df4.영어)]
df4
# 새로운 column을 생성하고 값은 lit comprehension으로 채워 넣는다.
```

```python
df5['국어학점'] = pd.Series()
df5['국어학점'] = df5['국어학점'].\
    mask(df5.국어  > 60, "F").mask(df5.국어 > 70, "C").mask(df5.국어 > 80, "B").mask(df5.국어 > 90, "A")
df5
# condition이 오름 차순으로 정렬되어야 한다.   🔰
```

```python
df = pd.DataFrame(np.random.randint(1,10, 14).reshape(2,7))
print(df.to_string())
# df.where(df > 4)
# condition에 맞는 것만 골라서 original value 보전, 나머진 A
df.mask(df > 4, '+4')
# condition에 맞는 것들 골라서 A
```

```python
s = pd.Series(range(0, 10, 1))
t = pd.Series([True, False, False, True, False, True])
```

```python
s.where(t, '*')
# 모자란 것은 True로 본다.
```

```python
s.mask(t, '*')
```

```python
s[s>4]
# condition에 맞는 놈만 추출, len(s)가 변경
```

```python
s.where(s > 4)
# condition에 맞는 놈만 original value를 보전, 나머지는 NaN, len(s)는 보존
```

```python
s.mask(s > 4, 'A')
# condition에 맞는 놈만 marking 'A', len(s)는 보존
```
