```python
# numpy.ufunc와 집계함수(aggregate(), agg())
# https://chat.openai.com/share/00bfdd25-40d9-4261-9a18-0fa117f9d59d
# map, apply, applymap 모두 argument로 func을 요구한다. 
import numpy as np 
import pandas as pd
```

```python
idx1 = ["송중기", "김나현", "권보아", "박효신"]
col1 = ["국어", "영어", "수학"]
val1 = [[82,81,77], [91,95,83], [78,72,88], [82,87,72]]
val2 = [["H", "T", "T"], ["T", "H", "H"], ["H", "H", "H"], ["T", "T", "T"]]
val3 = [list("HTTHHTHHT"), list("HTTHHTHTH"), list("HTTHTHTTT")]
df1 = pd.DataFrame(val1, index=idx1, columns=col1)
df2 = df1.copy()
df3 = pd.DataFrame(val2, index=idx1, columns=["1회", "2회", "3회"])
df4 = pd.DataFrame(val3, index=idx1[:3])
df4.columns = df4.columns + 1
a1 = list("ABCDEFG")
df = df1.copy()
df.loc["이지은"] = [81, 83, 91]
df["성별"] = list("남여여남여")
df["학년"] = list("21112")
```

```python
df
```

```python
df.pivot_table(index=['성별'], values=['국어'], columns=['학년'], aggfunc='mean' )
```

```python
df.pivot_table(index=['성별','학년'], aggfunc=['mean','median', 'std'] ).fillna('+')
```

```python
df = pd.DataFrame([[1, 2, 3],
                   [4, 5, 6],
                   [7, 8, 9],
                   [np.nan, np.nan, np.nan]],
                  columns=['A', 'B', 'C'])
df
```

```python
df.agg('sum', axis=0)
# string function name
```

```python
df.transform('sum')
# agg와 transform의 결정적인 차이   🔰🌸
```

```python
df.agg(['sum', 'min'], axis=1)
# list of function and/or function names
```

```python
df.agg({'A' : ['sum'], 'B' : ['min']})
# dicionnary
# column별로 함수를 지정하는 것은 dictionary로 하고
```

```python
df.agg(x=('A', 'max'), y=('B', 'min'), z=('C', np.mean))
# df.agg(x=(('A', 'B', 'C'), max))
# column과 그에 적용되는 함수가 1 대 1로 주어지며, index명을 따로 지정한다.   🔰
```
