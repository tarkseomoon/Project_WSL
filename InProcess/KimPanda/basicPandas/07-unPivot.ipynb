```python
import numpy as np
import pandas as pd
```

```python
val = [[83, 93, 99], [77, 88, 91], [92, 78, 95]]
idx = ['권보아', '김범수', '박효시']
df = pd.DataFrame(val, index=idx, columns=['국어', '수학', '과학'])

val1 = [[20, 10], [5, 25], [22, 8], [12, 18]]
idx1 = [['1반', '1반', '2반', '2반'], ['남', '여', '남', '여']]
df1 = pd.DataFrame(val1, index=idx1, columns=['도시락', '샌드위치'])
df2 = df1.unstack()
# stack, unstack의 default는 level = -1, 즉 index중에서 level이 가장 낮은 level, (0, 1) 중에 1에 대해서만 작용한다.   🔰

df3 = df1.copy()
df3['업체'] =['CJ푸드', '신세계', '신세계', 'CJ푸드']
# df4 = df3.reset_index().rename(columns={'level_0': '반', 'level_1':'성별'})
df4 = df3.reset_index(names=['반', '성별'])
#  reset_index는 deault는 모든 level에서 unstack한다.    🔰
df
```

```python
Unpivot_df = df.stack().reset_index()
# stack으로 unpivot하는 방법
# stack은 default level =1, reset_index는 모든 level   🔰
# DataFrame.reset_index(names=[...])는 가능하나, Series.reset_index()에는 names=[...]란 argument가 없다.   🔰
# Series.reset_index(name= " ")의 name argument는 original Series 값의 column name을 지정한다.   🔰
Unpivot_df.columns = ['이름', '과목', '점수']
Unpivot_df
```

```python
df_unpivot = df1.stack().reset_index()
df_unpivot.columns = ['반', '성별', '선택', '수량']
df_unpivot
```

```python
df2_unpivot.pivot(index='반', columns=['성별', '선택'], values='수량')
```
