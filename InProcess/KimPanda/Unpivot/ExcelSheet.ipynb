```python
import numpy as np 
import pandas as pd
import openpyxl
pd.options.display.max_rows = 30
pd.options.display.float_format = '{:,.2f}'.format
pd.set_option('expand_frame_repr', True)
# pd.describe_option()
# pd.reset_option()
```

```python
## openpyxl을 사용하여 특정 workbook에서 소속 sheet들의 이름 sheetname을 가져 온다.💮 
weekly = openpyxl.load_workbook('D02EXAMPLE.xlsx').sheetnames[:7]
```

```python
# upv_df = pd.DataFrame([])
# for i in weekly:
#   weekly_df = pd.read_excel('D02EXAMPLE.xlsx', sheet_name=i).set_index('제품')
#   upv_weekly_df = weekly_df.stack().reset_index()
#   upv_weekly_df.columns = ['제품', '지점', '매출']
#   upv_weekly_df['일자'] = i
#   upv_df = pd.concat([upv_df, upv_weekly_df] )

upv_df = pd.DataFrame([])
for i in weekly:
  weekly_df = pd.read_excel('D02EXAMPLE.xlsx', sheet_name=i)
  upv_weekly_df = pd.melt(weekly_df, id_vars='제품', value_vars=weekly_df.columns[1:], var_name='지점', value_name='매출')
  upv_weekly_df['일자'] = i
  upv_df = pd.concat([upv_df, upv_weekly_df] )
  
upv_df.head(7)
```

```python
branch_df = pd. read_excel('D02EXAMPLE.xlsx', sheet_name='지점구분')
inventory_df = pd. read_excel('D02EXAMPLE.xlsx', sheet_name='제품구분')
```

```python
upv_df = upv_df.merge(branch_df, how = 'left').merge(inventory_df, how = 'left')
upv_df.head(5)
```

```python
pvt_df = pd.pivot_table(upv_df, index='일자', columns='지역', values='매출', aggfunc='sum')
pvt_df
```

```python
pvt1_df = pd.pivot_table(upv_df, index='지역', columns='구분', values='매출', aggfunc='sum')
pvt1_df
```
