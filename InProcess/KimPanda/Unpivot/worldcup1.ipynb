```python
import numpy as np
import pandas as pd
# pd.options.display.max_rows = 20
pd.options.display.float_format = '{:,.2f}'.format
```

```python
worldcup_df = pd.read_excel('E14EXAMPLE.xlsx')
euros = pd.read_excel('E14EXAMPLE.xlsx', sheet_name=2)
worldcup_df
```

```python
# worldcup_df = worldcup_df.set_index('년도').iloc[: , 2:6]
# upv_worldcup_df =  worldcup_df.stack().reset_index()
# upv_worldcup_df.columns= ['년도', '순위', '국가']
# upv_worldcup_df

upv_worldcup_df = pd.melt(worldcup_df, id_vars='년도', value_vars=worldcup_df.columns[3:7], var_name='순위')
upv_worldcup_df.columns=['년도', '순위', '국가']
upv_worldcup_df
```

```python
pvt_worldcup_df = pd.pivot_table(upv_worldcup_df, index='국가', columns='순위', values='년도', aggfunc='count')
pvt_worldcup_df.sort_values(by=['1위','2위','3위','4위'], ascending=False, inplace=True)
pvt_worldcup_df[:8]
```

```python
euro_df = worldcup_df[worldcup_df['개최국'].isin(euros['국가'])]
upv_euro_df = pd.melt(euro_df, id_vars='년도', value_vars=euro_df.columns[3:7])
upv_euro_df.columns=['년도', '순위', '국가']
pvt_euro_df = pd.pivot_table(upv_euro_df, index='국가', columns='순위', values='년도', aggfunc='count')
pvt_euro_df.sort_values(['1위', '2위', '3위', '4위'], ascending=False, inplace=True)
pvt_euro_df[:8]
```
