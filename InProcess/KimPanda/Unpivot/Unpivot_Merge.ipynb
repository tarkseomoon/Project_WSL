```python
import pandas as pd
pd.options.display.max_rows=20
pd.options.display.float_format='{:.2f}'.format

df1 = pd.read_excel('D00EXAMPLE.xlsx')
df1
```

```python
# stack을 쓰는 경우

df1.set_index('노래', inplace = True)

upv1 = df1.stack().reset_index()

upv1.columns = ['심사자', '참가자', '점수']

upv1
```

upv1 = df1.melt(id_vars='노래', value_vars=('A', 'B', 'C', 'D', 'E'), var_name='참가자', value_name='점수')
upv1.rename(columns={'노래':'심사자'})

```python
df2 = pd.read_excel('D00EXAMPLE.xlsx', sheet_name=3)
df2
```

```python
upv1 = upv1.merge(df2, how = 'left')
upv1
```

```python
# pvt1 = upv1.pivot_table('점수', columns='성별', index='참가자', aggfunc=['mean', 'std'])
pvt1 = pd.pivot_table(upv1, values='점수', index='참가자', columns='성별', aggfunc=['median', 'std'])
pvt1
```
